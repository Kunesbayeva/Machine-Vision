```python
#Task3
from scipy import misc
import numpy as np
import matplotlib.image as img
import scipy.ndimage as ndimage

image  = misc.imread('imaage.jpg',mode="L")
noisy1 = image + 5 * image.std() * np.random.random(image.shape)
alot  = 2 * image.max() * np.random.random(image.shape)
noisy2 = image + alot
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(15,10))
plt.subplot(221)
plt.imshow(image,cmap = plt.get_cmap('gray'))
plt.subplot(222)
plt.imshow(noisy1,cmap = plt.get_cmap('gray'))
plt.subplot(223)
im_med = ndimage.median_filter(noisy2, 3)
plt.title('Median filter', fontsize=20)
plt.imshow(im_med, interpolation='nearest', vmin=0, vmax=5)
plt.imshow(image,cmap='gray')
plt.subplot(224)
plt.imshow((image-im_med),cmap='gray')


plt.show() 
```
