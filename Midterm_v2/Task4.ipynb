```python
import cv2
import os
import numpy as np
from skimage import data
import matplotlib.pyplot as plt
from skimage.data import data_dir
import matplotlib.image as pltimg
from skimage.morphology import disk
from skimage import feature

img = pltimg.imread('imaage.jpg')
im = data.imread(os.path.join(data_dir, "imaage.jpg"), as_grey=True)
#im = pltimg.imread('6_26.jpg')
temp = img[:,:,1]
img[:,:,1] = img[:,:,2]
img[:,:,2] = temp
# canny filter
edges1 = feature.canny(im)
edges2 = feature.canny(im, sigma=3)
fig = plt.figure(figsize=(15,10))
plt.subplot(221)
plt.title("Image", fontsize=15)
plt.imshow(im)
plt.subplot(222)
plt.title("Canny", fontsize=15)
plt.imshow(edges1, cmap=plt.cm.Reds)
plt.subplot(223)
plt.title("Sobel", fontsize=15)
plt.imshow(edges2, cmap=plt.cm.Reds)
plt.show()
```
